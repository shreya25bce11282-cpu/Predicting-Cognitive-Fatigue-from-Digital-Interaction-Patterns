```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)

# Ensure data directory exists
os.makedirs("/content/data/processed", exist_ok=True)



```

```python
n_sessions = 200

df = pd.DataFrame({
    "session_duration_min": np.random.randint(5, 120, n_sessions),
    "scroll_events": np.random.randint(50, 800, n_sessions),
    "interaction_density": np.random.uniform(1, 10, n_sessions),
    "late_night_usage": np.random.choice([0, 1], n_sessions, p=[0.4, 0.6]),
    "notification_exposure": np.random.randint(0, 30, n_sessions),
    "task_switches": np.random.randint(0, 20, n_sessions)
})

df["fatigue_score"] = (
    0.04 * df["session_duration_min"]
    + 0.002 * df["scroll_events"]
    + 0.6 * df["interaction_density"]
    + 1.5 * df["late_night_usage"]
    + 0.3 * df["notification_exposure"]
    + 0.4 * df["task_switches"]
    + np.random.normal(0, 1, n_sessions)
)
```

```python
# save csv there
df.to_csv("/content/data/processed/synthetic_fatigue_sessions.csv", index=False)
df.head()
```
